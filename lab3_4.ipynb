```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df.drop('MODEL', axis=1))
scaled_features = scaler.transform(df.drop('MODEL', axis=1))
scaled_df = pd.DataFrame(scaled_features, columns = df.drop('MODEL', axis=1).columns)
```

```python
from sklearn.model_selection import train_test_split
x = scaled_df
y = df['MODEL']
x_training_data, x_test_data, y_training_data, y_test_data = train_test_split(x, y, test_size = 0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
n_neighbors_k = 3
model = KNeighborsClassifier(n_neighbors = n_neighbors_k)
model.fit(x_training_data, y_training_data)
```

```python
predictions = model.predict(x_test_data)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
print(classification_report(y_test_data, predictions))
```

```python
print(confusion_matrix(y_test_data, predictions))
```

```python
from sklearn.metrics.cluster import contingency_matrix
contingency_matrix(y_test_data, predictions)
```

```python
error_rates = []
for i in np.arange(1, 101):
    new_model = KNeighborsClassifier(n_neighbors = i)
    new_model.fit(x_training_data, y_training_data)
    new_predictions = new_model.predict(x_test_data)
    error_rates.append(np.mean(new_predictions != y_test_data))
plt.plot(error_rates)
```

```python
#Генерируем матрицу ошибок
from sklearn.metrics import confusion_matrix
error_matrix = confusion_matrix(y_test_data, predictions)
print(error_matrix)
print()
print("Percentage of errors = " + str(round(1-error_matrix[1,1]/np.sum(error_matrix,axis=1).tolist()[1], 3)))
```

```python
#Выбор оптимального значения для K с помощью метода «Локтя»
from sklearn.neighbors import KNeighborsClassifier
error_rates = []
for i in np.arange(1, 101):
    new_model = KNeighborsClassifier(n_neighbors = i)
    new_model.fit(x_training_data, y_training_data)
    new_predictions = new_model.predict(x_test_data)
    error_rates.append(np.mean(new_predictions != y_test_data))
val, idx = min((val, idx) for (idx, val) in enumerate(error_rates))
plt.plot(error_rates)
print("K = ",idx)
print("Error = ",error_rates[idx])
```
